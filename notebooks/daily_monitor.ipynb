```python
def get_emo_sum(trade_date, df, pre_df):
    emo_sum = pd.DataFrame()
    emo_sum.loc[:, 'trade_date'] = [trade_date]
    emo_sum.set_index('trade_date', inplace=True)

    # 上涨股
    emo_sum.loc[trade_date, 'up_cnt'] = len(df[df.pct_chg>0])
    emo_sum.loc[trade_date, 'upstop_cnt'] = len(df[df.limit=='U'])
    
    # 阳线股
    emo_sum.loc[trade_date, 'yang_cnt'] = len(df[df.bar_type=='yang'])
    pre_yas = pre_df[pre_df.bar_type=='yang'].index
    pre_yas_today = df[df.index.isin(pre_yas)]
    emo_sum.loc[trade_date, 'pre_ya_avg'] = pre_yas_today.pct_chg.mean()
    
    # 连涨股数
    emo_sum.loc[trade_date, 'upstops_cnt'] = len(df[df.conseq_up_num>=2])
    emo_sum.loc[trade_date, 'ups_cnt'] = len(df[(df.pct_chg>0) & (df.pre_pct_chg>0)])
    
    # 昨日上涨的票今日走势
    pre_ups = pre_df[pre_df.pct_chg>0].index
    pre_ups_today = df[df.index.isin(pre_ups)]
    emo_sum.loc[trade_date, 'pre_up_avg'] = pre_ups_today.pct_chg.mean()
    emo_sum.loc[trade_date, 'pre_up_cvo_avg'] = (pre_ups_today.pct_chg - pre_ups_today.open_pct).mean()
    
    # 昨日下跌的票今日走势
    pre_dns = pre_df[pre_df.pct_chg<0].index
    pre_dns_today = df[df.index.isin(pre_dns)]
    emo_sum.loc[trade_date, 'pre_dn_avg'] = pre_dns_today.pct_chg.mean()
    emo_sum.loc[trade_date, 'pre_dn_cvo_avg'] = (pre_dns_today.pct_chg - pre_dns_today.open_pct).mean()

    emo_sum.loc[trade_date, 'dn_cnt'] = len(df[df.pct_chg<0])
    emo_sum.loc[trade_date, 'dnstop_cnt'] = len(df[df.limit=='D'])
    emo_sum.loc[trade_date, 'fl_cnt'] = len(df[df.pct_chg==0])
    
    emo_sum.loc[trade_date, 'pct_avg'] = df.pct_chg.mean()
    
    # 赚钱效应
    
    
    return emo_sum


emo_sums = pd.DataFrame()
p_df = t_df = None

for i, trade_date in enumerate(tdu.past_trade_days()[-10:]):
    if t_df is not None:
        p_df = t_df
    t_df = df.xs(trade_date, level='trade_date', drop_level=True)
    if p_df is not None:
        emo_sum = get_emo_sum(trade_date, t_df, p_df)
        emo_sums = emo_sums.append(emo_sum)

emo_sums
```
