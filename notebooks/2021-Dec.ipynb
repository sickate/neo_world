### 3. Lessons Learned

#### 炸板股的选择要点
* 不能放量，上方不能有大量压力
* 最好接近破前高或者创新高
* 上午触板的比下午触板的好，封板时间较长的好，长期在板上封不住又总想封的不好
* 炸板后仍然维持比封板前低位高的（日内阳线），如 1119 清源股份
* 1117：昨天的炸板股涨得好的：
  * 山东威达：炸板日【个股突破半年新高，放量，下午上板。工业母鸡板块首阴】，次日板块大涨个股涨停
  * 宝胜股份：炸板日【下午上板，个股放量突破新高，下午上板，线缆板块大涨后揉搓线】，次日板块大涨个股涨停
  * 永鼎股份：炸板日【个股突破年内新高，上午上板，封板时间短，通信线缆板块高位盘整】，次日板块大涨个股涨停
* 1118：昨天的炸板股涨得好的（**次日大盘大跌**）
  * 生意宝：炸板日【新高，下午上板，封板时间较长】大涨，无板块
  * 拓邦股份：炸板日【接近前高，下午上板，封板时间较长】次日冲高接近封板回落，无板块

#### 大活跃的要点
* 实际换手率太大的不好，最好在 15% 以内，5%以上
* 最近的一波冲高不能太高，否则抛压重
* 1119：一般超预期和超级超预期的区别
  * 盾安环境（深V捞起，有点超预期，很可能反包）：小阳
  * 萃华珠宝（核按钮捞起，日内几乎没有亏钱的）：长上影十字
  * 盛洋科技（超预期，明日在突破前高后可跟）：一度涨停
  * 1、2票次日（情绪指标53%）均大涨，3 冲高长上影

#### 情绪的要点
* 大涨票，最好能在大盘弱势的时候，缩量回调均线，此时介入盈亏比较高
* 追高的总结：前面大涨的票，次日有几种情况
  * 高开上涨，随后下跌，然后回拉均线并缠绕，偏离均线向下的幅度越来越小，最后翻上去，此时可以买
  * 不能在均线以下抄底，风险太大。抄底必须是连续下跌后的急跌中
  * 300 的票次日大概率要下穿分时均线下方，不穿不追高。穿了不翻上去也不能追。理想情况下次日分时均线保持在7%以上
  * 有的时候次日会走小幅冲高-》走弱大幅跌破分时-》日内慢慢再度翻上分时，此时介入胜率较高

#### 斐波那契空间
* 连续上涨的票见顶下跌后，从底部到顶点回调到 50%时通常有一个反抽机会

#### 急杀低吸的要点
* 待观察

#### 日内走弱卖出的要点
* 高开低走的，回调均线再回落，果断卖出
* 如果回调均线不准备卖，就坚决等反弹，或者破位（日内跌破 5%以上且日线跌破 ma5）后再卖。不能情绪化杀跌！

#### 冲高卖出要点
* 早盘急拉高的可速卖：300 的9%回1.5%，主板的7%回1%
* 11 点以后可以把条件单撤了，这个时候拉高的日内不回撤的可能性较大

```python
def sum_diff(df_ori, df_filtered):
    len_ori = len(df_ori)
    len_filtered = len(df_filtered)
    result = {}
    result['next_open_pct_mean'] = {'Original': df_ori.next_open_pct.mean(), 'Filtered': df_filtered.next_open_pct.mean()}
    result['next_pct_chg_mean'] = {'Original': df_ori.next_pct_chg.mean(), 'Filtered': df_filtered.next_pct_chg.mean()}
    result['next2_pct_chg_mean'] = {'Original': df_ori.next2_pct_chg.mean(), 'Filtered': df_filtered.next2_pct_chg.mean()}
    result['next3_pct_chg_mean'] = {'Original': df_ori.next3_pct_chg.mean(), 'Filtered': df_filtered.next3_pct_chg.mean()}
    result['next_up_rate'] = {
        'Original': len(df_ori[df_ori.next_pct_chg>0])/len_ori, 
        'Filtered': len(df_filtered[df_filtered.next_pct_chg>0])/len_filtered
    }
    result['next2_up_rate'] = {
        'Original': len(df_ori[df_ori.next2_pct_chg>0])/len_ori, 
        'Filtered': len(df_filtered[df_filtered.next2_pct_chg>0])/len_filtered
    }
    result['next3_up_rate'] = {
        'Original': len(df_ori[df_ori.next3_pct_chg>0])/len_ori, 
        'Filtered': len(df_filtered[df_filtered.next3_pct_chg>0])/len_filtered
    }
    return result
    
def back_test(strat, df, columns=None, limit=10):
    res, keys = strat.get_result(df)
    res_dict = sum_diff(df, res)
    display(pd.DataFrame(data=res_dict))
    if not columns:
        columns = list(set(['name'] + keys + ['vol_ratio', 'turnover_rate_f', 'next_pct_chg', 'next2_pct_chg']))
    res1 = res.sort_index(level='trade_date').tail(limit)[columns]
    display(style_df(res1))
    return res

df.loc[:, 'pre2_pct_chg'] = df.groupby('ts_code').pct_chg.shift(2)
df.loc[:, 'pre2_conseq_up_num'] = df.groupby('ts_code').conseq_up_num.shift(2)
df.loc[:, 'pre2_close'] = df.groupby('ts_code').close.shift(2)
```

```python
%%time

# 计算最近5天涨幅大于7%的天数
df.loc[:,'pre5_big_up'] = df.groupby('ts_code').rolling(5).pct_chg.apply(lambda x: (x>=7).sum()).reset_index(level=0, drop=True) 
```

```python
# 创 

t1 = Strategy('chuang', StockFilter(end_date).cyb().not_st())
t1.add_condition('close', '>', var='ma_close_5')
t1.add_condition('close', '<', var='ma_close_5', ratio=1.05)
t1.add_condition('ma_close_5', '>', var='ma_close_10')
t1.add_condition('vol_ratio', '<', 0.7)
t1.add_condition('turnover_rate_f', '>', 5)
t1.add_condition('turnover_rate_f', '<', 25)
t1.add_condition('pre5_big_up', '>=', 1)

res = back_test(df=df, strat=t1, limit=20, columns=['name', 'pre3_pct_chg', 'pct_chg', 'vol_ratio', 'turnover_rate_f', 'next_open_pct', 'next_pct_chg', 'next2_pct_chg', 'next3_pct_chg'])
```
